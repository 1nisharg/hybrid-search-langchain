```python
from langchain_community.retrievers import PineconeHybridSearchRetriever
```

```python
import os
from pinecone import Pinecone, ServerlessSpec
index_name="pinecone-hybrid"


#init the pinecone client
pc=Pinecone(api_key=api_key)

#create index
if index_name not in pc.list_indexes().names():
  pc.create_index(
      name=index_name,
      dimension=384, #dimension of dense vector
      metric='dotproduct', ##sparse values supported only for dotproduct
      spec=ServerlessSpec(cloud='aws', region='us-east-1')
  )

```

```python
index=pc.Index(index_name)
index
```

```python
!pip install langchain-huggingface
```

```python
##vector embedding and sparse matrix
import os
from dotenv import load_dotenv
load_dotenv()


from langchain_huggingface import HuggingFaceEmbeddings
embeddings=HuggingFaceEmbeddings(model_name='all-MiniLM-L6-v2')
```

```python
!pip install pinecone-text
```

```python
from pinecone_text.sparse import BM25Encoder

bm25_encoder=BM25Encoder().default() #responsible for sparse matrix
bm25_encoder
```

```python
sentences=[
    "In april, I started OpenRAG",
    "In may, we built our prototype",
    "in june, we conducted PAN india hackathon",
]

#tfidf values on these sentences
bm25_encoder.fit(sentences)

#store the values to a json
bm25_encoder.dump("bm25_values.json")
```

```python
retriever=PineconeHybridSearchRetriever(embeddings=embeddings, sparse_encoder=bm25_encoder, index=index)
```

```python
retriever
```

```python
retriever.add_texts(
    [
    "In april, I started OpenRAG",
    "In may, we built our prototype",
    "in june, we conducted PAN india hackathon",
]

)
```

```python
retriever.invoke("What did I did in april?")
```
